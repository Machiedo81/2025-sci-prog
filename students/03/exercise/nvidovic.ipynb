```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
employee_productivity_df = pd.read_csv("/workspaces/2025-sci-prog/students/03/data/13_employee_productivity.csv")

print("Head: \n", employee_productivity_df.head())
```

```python
slope, intercept, r, p, std_err = stats.linregress(employee_productivity_df['years_experience'], employee_productivity_df['productivity'])
```

```python
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, employee_productivity_df['years_experience']))

plt.scatter(employee_productivity_df['years_experience'], employee_productivity_df['productivity'])
plt.plot(employee_productivity_df['years_experience'], mymodel)
plt.show()
```
