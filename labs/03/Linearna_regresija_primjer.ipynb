## Linearna regresija u Pythonu korištenjem NumPy biblioteke

Python ima metode za pronalaženje odnosa između točaka podataka i za crtanje pravca linearne regresije. Ispod će kroz primjer biti objašnjeno kako koristiti ove metode bez objašnjenja matematičkih formula tj. kako se matematičkim putem dolazi do koeficijenata a i b u linearnoj regresiji.

Započet ćemo sa crtanjem raspršenih točaka na ravni, koje su definirane listom x koordinata i listom y koordinata.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(range(1,14))
y = np.array([2,8,8,18,25,21,32,44,32,48,61,45,62])

plt.scatter (x,y)

plt.show()
```

Sada ćemo odrediti pravac linearne regresije koji se najbolje uklapa s točkama podataka.

U ovom primjeru koristimo NumPy nizove koji su korisniji od Python listi jer se mogu tretirati kao vektori nad kojima se mogu izvršiti operacije (npr. zbrajanje vektora, množenje vektora i skalara itd.) koje značajno pojednostavljuju računanja.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(range(1,14))
y = np.array([2,8,8,18,25,21,32,44,32,48,61,45,62])

[a,b]=np.polyfit(x,y,1)
print(a)
print(b)

plt.scatter (x,y)
plt.plot ([0,15],[b,a*15+b])
plt.show()
```

## Točnost modela 
Određivanje koeficijenta determinacije $\ R^2  $  
$\ R^2  $ - je vrijednost u rasponu od 0 do 1, što je bliže 1 to je model linearne regresije točniji.

```python
from sklearn.metrics import r2_score
import numpy as np
from numpy import mean

x = np.array(range(1,14))
y = np.array([2,8,8,18,25,21,32,44,32,48,61,45,62])
[a,b]=np.polyfit(x,y,1)

r2_score(y, a*x+b)
```

## Predviđanje vrijednosti pomoću linearne regresije

```python
import matplotlib.pyplot as plt

plt.scatter (x,y)
plt.scatter (20, a*20+b, color ='red')
plt.plot ([0,25],[b,25*a+b])
plt.title('Graf prikazuje predviđenu vrijednost upotrebom linearne regresije')
plt.show ()

print (a*20+b)
```
